```python
import matplotlib.pyplot as plt
import numpy as np
```

<h2>Soal 1: Multiple Subplots</h2>

Dengan data di bawah ini buatlah visualisasi seperti expected output:

```python
x = np.linspace(2*-np.pi, 2*np.pi, 200)
tan = np.tan(x)/10
cos = np.cos(x)
sin = np.sin(x)
```

```python
# code here
# Membuat figure dan axes
fig, ax = plt.subplots(nrows=1, ncols=3, sharey=True, figsize=(12,8))
ax = ax.flatten()

# Memasukkan data ke setiap axes dan membuat title plot
ax[0].plot(x,tan, label='tanx', color='blue')
ax[0].set_title('Grafik tanx')

ax[1].plot(x,sin, label='sinx', color='red')
ax[1].set_title('Grafik sinx')

ax[2].plot(x,cos, label='cosx', color='green')
ax[2].set_title('Grafik cosx')

# Membuat legend pada figure
fig.legend(loc=9, bbox_to_anchor=(0.,1.,1.,0.102))

# Membuat jarak antar plot
plt.tight_layout(1,2)

# Menampilkan plot
plt.show()
```

<h2>Soal 2: Nested Axis</h2>

Dengan data di bawah ini, buatlah visualisasi seperti expected output:

```python
x = np.linspace(2*-np.pi, 2*np.pi, 100)
y = np.cos(x)
y2 = np.cos(x**2)
y3 = np.cos(x**3)
y4 = np.cos(x**4)
y5 = np.cos(x**5)
```

```python
# code here
# Membuat figure dan Axes
fig, ax = plt.subplots(figsize=(12,8))

# Memasukkan data ke dalam Axes Utama
ax.plot(x,y)

# Memasukkan data ke dalam Sub Axes Pertama
ins1 = ax.inset_axes([0.9,0.9,0.3,0.3], sharex=ax)
ins1.plot(x,y, color='blue')
ins1.plot(x,y4,color='black')
ins1.set_title('Grafik CosX^4')

# Memasukkan data ke dalam Sub Axes Kedua
ins1 = ax.inset_axes([-0.1,0.9,0.3,0.3], sharex=ax)
ins1.plot(x,y,color='blue')
ins1.plot(x,y5, color='yellow')
ins1.set_title('Grafik CosX^5')

# Memasukkan data ke dalam Sub Axes Ketiga
ins1 = ax.inset_axes([-0.1,-0.1,0.3,0.3], sharex=ax)
ins1.plot(x,y2,color='red')
ins1.plot(x,y, color='blue')
ins1.set_title('Grafik CosX^2')

# Memasukkan data ke dalam Sub Axes Keempat
ins1 = ax.inset_axes([0.9,-0.1,0.3,0.3], sharex=ax)
ins1.plot(x,y3,color='green')
ins1.plot(x,y, color='blue')
ins1.set_title('Grafik CosX^2')

# Membuat judul/title pada figure
fig.suptitle('Grafik CosX')

# Menampilkan plot
plt.show()
```
