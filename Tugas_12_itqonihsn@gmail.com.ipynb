Dengan menggunakan data di bawah ini, plot kan data tersebut kemudian custom sehingga memiliki result seperti yang di harapkan :

note: warna terdiri dari warna Hijau, kuning, Biru, dan Merah

```python
x = np.linspace(-10, 10, 100)
cos = np.cos(x)
cos3 = np.cos(x)*3
sin = np.sin(x-90)
sin2 = np.sin(x)*2
```

```python
# code here
# Import library
import numpy as np
import matplotlib.pyplot as plt

# Membuat komponen figure dan axes 
fig, ax = plt.subplots(figsize=(12,8))

# Membuat axes label x dan y
ax.set_xlabel('X-Axis Label')
ax.set_ylabel('Y-Axis Label')

# Membuat judul/title grafik
ax.set_title('Grafik Cos Sin')

# Memasukkan data kedalam axes
ax.plot(x,cos, ls='-', label='Cosx', color='red')
ax.plot(x,cos3, ls='--', label='3 Cosx', color='green')
ax.plot(x,sin, ls='-.', label='Sin(x-90)', color='blue')
ax.plot(x,sin2, ls='None', marker='x', label='2 Sinx', color='yellow')

# Membuat legend 
plt.legend(loc=10, framealpha=0)

# Menampilkan grafik
plt.show()
```
