```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv('vgsales.csv')
df.head()
```

<h2>Soal 1: Preparing Data</h2>

Manipulasi data tersebut dengan menggroupkan berdasarkan Genre, kemudian ambil rata2 penjualan untuk setiap Region Kecuali Global_Sales berdasarkan kategori Genre.

```python
#code here
df_group = df.groupby('Genre')[['NA_Sales','EU_Sales',"JP_Sales",'Other_Sales']].mean()
df_group
```

<h2>Soal 2: Perbandingan Kuantitatif Barplot - Grouping Visualisasi Dengan Barplot</h2>

1. Visualisasikan gambar yang tadi kita manipulasi, untuk menghasilkan visualisasi seperti di bawah ini
2. Tuliskan apa insight yang bisa kamu dapat dari visualisasi tersebut

```python
# code here

# Membuat figure dan axes
fig, ax = plt.subplots(figsize=(12,8))

# Mendefinikan x dan width sebagai posisi 
x = np.linspace(0,14,4)
width = 0.3

# Looping memasukkan data menjadi plot ke dalam axes
for i,j in enumerate(df_group.index):
    ax.bar(x+(width*i), df_group.iloc[i], width, label=j)

# Membuat kategori pada sumbu X    
plt.xticks(x+1.8, df_group.columns)

# Memberikan penamaan pada sumbu X, Y dan Judul Axes
ax.set_xlabel('Region Sales')
ax.set_ylabel('Mean Sales')
ax.set_title('Mean Sales Video Games By Genre')

# Mengedit Axes
ax.set_axisbelow(True)
ax.set_facecolor(color='0.9')

# Membuat grid
plt.grid(color='white')

# Membuat legend
fig.legend()

# Menampilkan plot
plt.show()
```

Jawab: 
Dari visualisasi di atas kita dapat mengambil insight:
- Pada grafik barplot, dibandingkan dengan pie plot, salah satu kelebihan barplot adalah kita dapat melihat selisih antar nilai/frekuensi variabel dengan jelas.
- Dari keempat region/country yang dibandingkan, rata-rata penjualan video games paling tinggi berada di North America pada setiap genrenya. sedangkan rata-rata terendah berada di region/country lainnya. Hal ini dapat disimpulkan bahwa di North America terdapat banyak sekali pecinta game / gamers dibandingkan dengan region/country lain. Pasar di dunia game pada North America sangatlah tinggi dan signifikan.
- Dari keempat region, genre platform dan shooter cenderung memiliki rata-rata penjualan yang tinggi di setiap regionnya, sedangkan genre adventure memiliki rata-rata penjualan yang rendah di setiap regionnya.
- Pada grafik North America dan europe, genre video games tertinggi yaitu platform dan disusul oleh shooter, dan terendah yaitu genre adventure.
- Sedangkan pada grafik Japan, genre video games tertinggi yaitu Role-Playing dan yang terendah adalah shooter. Hal ini dapat disimpulkan bahwasanya kesukaan para gamers terhadap genre video games shooter di japan berbanding terbalik dengan para gamers di NA dan EU.
- Pada grafik Other_Sales, saya rasa kurang dapat diambil insightnya karena tidak jelas/spesifik other ini terdiri dari berapa country/region, mungkin gabungan dari 5 country atau mungkin lebih. Yang jelas dibandingkan dengan penjualan di region NA, EU, dan JAPAN, rata-rata penjualan selain di region tersebut masih sangat rendah.

<h2>Soal 3: Perbandingan Kuantitatif Barplot - Stack Barplot</h2>

1. Visualisasikan gambar yang tadi kita manipulasi, untuk menhasilkan visualisasi seperti di bawah ini
2. Tuliskan apa insight yang bisa kamu dapat dari visualisasi tersebut

```python
#code here

# Membuat figure dan axes
fig, ax = plt.subplots(figsize=(12,8))

# Mendefinikan x dan width sebagai posisi 
x = np.linspace(0,14,4)
width = 1.3
y_bottom = 9
x = np.linspace(0,5,4)

# Looping memasukkan data menjadi plot ke dalam axes
for i,j in enumerate(df_group.index):
    ax.bar(x, df_group.iloc[i], width, label=j, bottom=y_bottom)
    y_bottom += df_group.iloc[i]

# Membuat kategori pada sumbu X   
plt.xticks(x, df_group.columns)

# Memberikan penamaan pada sumbu X, Y dan Judul Axes
ax.set_xlabel('Region Sales')
ax.set_ylabel('Mean Sales')
ax.set_title('Mean Sales Video Games By Genre')

# Mengedit Axes
ax.set_axisbelow(True)
ax.set_facecolor(color='0.9')

# Membuat grid
plt.grid(color='white')

# Membuat legend
fig.legend()

# Menampilkan plot
plt.show()
```

Jawab:
- Perbedaan grafik barplot dan stacked barplot, pada grafik stacked barplot kita dapat melihat perbandingan nilai frekuensi total variabel dalam setiap variable/index X. Dimana pada grafik di atas dapat dilihat total rata-rata penjualan tertinggi berapa di NA_Sales sebesar 12.0 lebih, disusul dengan EU_Sales sekitar 10.7, lalu JP_Sales sekitar 10.1 dan rata-rata penjualan terendah berada di Other_Sales sekitar 9.6. 
- Rata-rata penjualan video games lebih besar dibandingkan yang lain berdasarkan genrenya, hal ini dapat dilihat dari besarnya volume bar pada setiap genrenya. 
- Pada grafik stacked barplot di atas, dapat dilihat juga perbandingan genre video games antar region/country nya. Misal pada genre video games platform, rata-rata penjualan di NA_Sales lebih dibandingkan dengan EU,JP,Other Sales.
