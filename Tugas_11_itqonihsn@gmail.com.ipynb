- Axes adalah area dalam figure dimana data akan dibuat plot. Berupa grafik dimana dalam satu figure bisa terdapat beberapa Axes.

<h2>Soal 2: Membuat Component Figure dan Axis</h2>

- Buatlah Figure dan Axis kosong tanpa Data
- Plot data di bawah ini dengan 2 cara, yaitu dengan membuat figure dan axis secara explicit dan Implicit

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 20, 100)
y = np.cos(x/2)
```

```python
# Membuat figure dan Axes kosong
fig = plt.figure()
ax = fig.add_subplot();

# Membuat plot dengan cara explicit
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x,y);
```

```python
# Membuat plot dengan cara implicit
plt.plot(x,y);
```

<h2>Soal 3: Memasukan multiple data ke dalam satu axis</h2>

- Plot ketiga data berikut ke dalam satu axis, dan gunakan seaborn style
- save data tersebut dengan nama file bebas (apa saja) dan tampilkan kembali gambar tersebut.

```python
x = np.linspace(0, 20, 100)
y = np.cos(x/2)
y2 = np.sin(x)
y3 = np.sin(2*x)
```

```python
# Membuat figure dan axes
fig = plt.figure(dpi=300)

ax = fig.add_axes([0,0,1,1])

plt.style.use('seaborn')

ax.plot(x,y)
ax.plot(x,y2)
ax.plot(x,y3)

plt.show()
```
