```python
# Import library
from matplotlib import pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
# Meload datase
df = sns.load_dataset('diamonds')
df.head()
```

# Soal 1: Scatter Plot

Buatlah scatterplot menggunakan Seaborn dari data df di atas, variasikan menggunakan palette yang telah disediakan sehingga didapatkan output sesuai expected.

```python
# Mengatur ukuran figure
sns.set({'figure.figsize':(13,13)})

# Mendefinisakan palette warna yang akan dipakai
palette = "ch:r=-.2,d=.3_r"

# Membuat scatter plot 
sns.scatterplot(x='carat', y='price', data=df, hue='clarity', size='depth', linewidth=0, sizes=(10,20), palette=palette)

# Menampilkan plot
plt.show()
```

# Soal 2: Catplot

Buatlah sebuah Categorical plot dengan kind = 'boxplot', sehingga diperoleh hasil seperti expected output berikut:


```python
# Membuat categorical plots dengan style tertentu dan tanpa menampilkan nilai outlier
with sns.axes_style(style='ticks'):
    sns.catplot(x='clarity', y='price', data=df, col='cut', kind='box', height=8, showfliers=False)

# Menampilan plot
plt.show()
```

Expected Output:

![alt text](https://drive.google.com/uc?id=17mSiL-qQtFB2D8rUYB0v7d6tOrOHDoh6)

# Soal 3: Pair Plot

Ambil 300 data secara acak dari df, kemudian buat pair plotnya dari 300 data tersebut untuk kolom carat, depth, table, dan price serta markers =  ['p', 'o', 's', 'D', 'X'] sehingga dihasilkan output sesuai expected.

```python
# Membuat300 data acak dari dataset df
df = df.sample(n=300, random_state=123)
df.head()

# Membuat pair plot menjadi matrix plot segitiga bawah
sns.pairplot(df, vars=['carat','depth','table','price'], markers=['p','o','s','D','X'], hue='cut', corner=True)

# Menampilkan plot
plt.show()
```

Expected Output:

![alt text](https://drive.google.com/uc?id=12rStINwUNGpqZd1iIGaqzTIvw9irAkJ0)

# Soal 4: Joint Plot

Gunakan data yang sama dengan soal 3 untuk menghasilkan dua visualisasi joint plot antara carat dan price nya sesuai expected output.

```python
# Membuat joint plot dalam bentuk kernel density estimation (kde) dan garis regresi
with sns.axes_style('white'):
    sns.jointplot(x='carat', y='price', data=df, kind='kde')
    sns.jointplot(x='carat', y='price', data=df, kind='reg')

# Menampilkan plot
plt.show()
```
